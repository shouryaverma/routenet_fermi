```python
import os

def extract_loss_values(folder_path):
    loss_values = []
    filenames_and_losses = []
    
    for filename in os.listdir(folder_path):
        if '.data' in filename:
            parts = filename.split('-', 1)
            if len(parts) > 1:
                first_num = int(parts[0])
                loss = float(parts[1].split('.data')[0])
                filenames_and_losses.append((first_num, loss))
    
    # Sort by the first number
    sorted_values = sorted(filenames_and_losses, key=lambda x: x[0])
    
    # Extract just the loss values after sorting
    loss_values = [loss for _, loss in sorted_values]
    
    return loss_values
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data for Mixed Traffic Models with Scheduling Policies
scheduling_data = {
    'Metric': ['Delay MAPE', 'Jitter MAPE', 'Losses MAE'] * 4,
    'Model': (['RNN Model'] * 3 + ['LSTM Model'] * 3 + 
              ['GRU Model'] * 3 + ['Paper GRU'] * 3),
    'Value': [
        # RNN Model
        9.80, 29.70, 0.7464,
        # LSTM Model
        2.96, 16.73, 0.2218,
        # GRU Model
        3.82, 17.01, 0.2063,
        # Paper GRU
        3.35, 17.21, 0.1978
    ]
}

# Create DataFrame
df = pd.DataFrame(scheduling_data)

# Set up the plot
plt.figure(figsize=(8, 6), dpi=300)

# Create grouped bar plot
x = np.arange(len(df['Metric'].unique()))
width = 0.2
multiplier = 0

# Plot bars for each model
for model in df['Model'].unique():
    model_data = df[df['Model'] == model]['Value']
    offset = width * multiplier
    plt.bar(x + offset, model_data, width, label=model)
    multiplier += 1

# Customize the plot
plt.xlabel('Metric')
plt.ylabel('Value (%)')
plt.title('Model Performance Comparison with Different Scheduling Policies')
plt.xticks(x + width * 1.5, df['Metric'].unique())
plt.legend(loc='upper left')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of each bar, using different formats for MAPE and MAE
for idx, container in enumerate(plt.gca().containers):
    labels = []
    for val in container:
        value = val.get_height()
        if value < 1:  # For MAE values
            labels.append(f'{value:.6f}')
        else:  # For MAPE values
            labels.append(f'{value:.2f}%')
    plt.bar_label(container, labels=labels, padding=3, size=8)

# Set y-axis to log scale since values vary significantly
plt.yscale('log')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```
